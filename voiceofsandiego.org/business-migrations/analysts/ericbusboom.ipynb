```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sys 
```

```python
# row labels ( index) are the source CBSA, and the column labels are the destination
df = pd.read_csv('/Volumes/Storage/AmbryWarehouse/cabusmigrations/camigrations.csv')
```

```python
# dir == 0 means business stayed in state, 1 == moved in, -1 == moved out
df['dir'] = df.apply(lambda x: (int(x.originstate == 'CA') * -1 ) + (int(x.deststate == 'CA')), axis = 1 )
df.count()
```

```python
# Omit all records for businesses that stayed in the same CBSA
df = df[df.origincbsa != df.destcbsa ]
df.count()
```

```python
df.columns
```

```python
odf = df[['moveyear','industry','originstate','origincbsa','origin_metro_area','jobs','estabs']]
odf.columns = ['year','industry','state','cbsa','metro','jobs','estabs']
odf['sig_jobs'] = -odf['jobs']
odf['sig_estabs'] = -odf['estabs']
idf = df[['moveyear','industry','deststate','destcbsa','destination_metro_area','jobs','estabs']]
idf.columns = ['year','industry','state','cbsa','metro','jobs','estabs']
idf['sig_jobs'] = idf['jobs']
idf['sig_estabs'] = idf['estabs']

hdf = odf.append(idf, ignore_index = True)
hdf.head()
```

```python
hdf[hdf.cbsa == 41740].groupby('year').sum().sig_jobs.plot()
```

```python
df['sig_jobs'] = df['dir'] * df['jobs']
df['sig_estabs'] = df['dir'] * df['estabs']
```

```python
df.groupby('moveyear').sum()
```

```python
sums = df.groupby('moveyear').sum()
jobs_sums = sums['sig_jobs'].cumsum()
jobs_sums.plot()
```

```python
estab_sums = sums['sig_estabs'].cumsum()
estab_sums.plot()
```

```python
sd_sums = df[(df.origincbsa == 41740 ) | (df.destcbsa == 41740 )].groupby('moveyear').sum()
sd_sums['sig_jobs'].cumsum().plot()
```
