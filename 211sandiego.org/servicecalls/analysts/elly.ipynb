```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# You'll probably need to run 'sudo pip install sdipylib' for this to work
from  sdipylib.url import cache_url
pd.__version__
```

```python
fn = cache_url('http://devtest.sandiegodata.org.s3.amazonaws.com/manifests/fc22cee6-acfa-4236-a6cb-e0d887eba178/211_service_calls.csv')
df = pd.read_csv(fn,dtype =  {16: str, 26: str, 27: str, 29: str})
```

```python
cat_need_df = df[['need_tax_cat','need_unmet']]
cat_need_df.dropna()
cat_need_df['met'] = 0
cat_need_df['unmet'] = 0
cat_need_df.met = cat_need_df.need_unmet.apply(lambda e: 1 if e == 'f' else 0)
cat_need_df.unmet = cat_need_df.need_unmet.apply(lambda e: 1 if e == 't' else 0)
cat_need_df = cat_need_df.drop("need_unmet", axis = 1)
```

```python
grouped = cat_need_df.groupby('need_tax_cat')
grouped = grouped.sum()
grouped = grouped.sort(['met'], ascending = [0])
grouped = grouped[grouped.unmet > 50]
grouped = grouped[grouped.met < 15000]
grouped.rename(columns={0:'Categories', 1:'Needs Met', 2:'Needs Unmet'}, inplace=True)
```

```python
grouped.plot(kind='barh', position=0.5, stacked=True)
```
