## Daily and Monthly Call Volume 2013-2014

```python
service_call_data["date"] = service_call_data.create_time.apply(lambda d : d.date())
call_volume_per_day = service_call_data.groupby(['date']).size()

print "Day of least calls is Christmas {0}, and day of most calls is {1}".format(call_volume_per_day.argmin(), call_volume_per_day.argmax())
daily_call_volume_summary = call_volume_per_day.describe()
daily_call_volume_summary
```

The average number of daily calls is roughly <b>2451</b>,<br />
and the least number of calls occurs on <b>December 25th, 2013</b>,<br />
while the most number of calls occurs on <b>March 24th, 2014</b>

```python
start_2013 = call_volume_per_day.index.searchsorted(dt.date(2013, 1, 1))
end_2013 = call_volume_per_day.index.searchsorted(dt.date(2013, 12, 31))
start_2014 = call_volume_per_day.index.searchsorted(dt.date(2014, 1, 1))
end_2014 = call_volume_per_day.index.searchsorted(dt.date(2014, 12, 31))

fig, axes = plt.subplots(nrows=2, ncols=1); ylim = [0,call_volume_per_day.max()]
axes[0].set_title('Jul-Dec 2013'); axes[0].axhline(y=daily_call_volume_summary["mean"], color='r')
axes[1].set_title('Jan-Jun 2014'); axes[1].axhline(y=daily_call_volume_summary["mean"], color='r')

daily_call_volume_2013 = call_volume_per_day[start_2013:end_2013].plot(ax=axes[0], figsize=(18,6), color='b', ylim=ylim);
daily_call_volume_2014 = call_volume_per_day[start_2014:end_2014].plot(ax=axes[1], figsize=(18,6), color='g', ylim=ylim);
```

One would expect a surge in calls in the week of May 2014, during the San Diego wildfires, but even so the maximum amount of calls occurred on the 3rd to 4th week of March 2014.

```python
service_call_data["hour"] = service_call_data.create_time.apply(lambda d : d.hour)
call_volume_and_hour_bins_vs_counts = service_call_data[["date", "hour"]].groupby(["date", "hour"]).size()
call_volume_vs_hour_bins = call_volume_and_hour_bins_vs_counts.unstack(level=1).fillna(0)
call_volume_vs_hour_bins.index = pd.to_datetime(call_volume_vs_hour_bins.index)
call_volume_vs_hour_bins_on_month = [call_volume_vs_hour_bins[call_volume_vs_hour_bins.index.month == m] for m in range(1,len(month_abbr))]
```

```python
service_call_data["dayofweek"] = service_call_data.create_time.apply(lambda d : day_abbr[d.weekday()])
dayofweek_and_hour_bins_vs_counts = service_call_data[["dayofweek", "hour"]].groupby(["dayofweek", "hour"]).size()
dayofweek_vs_hour_bins = dayofweek_and_hour_bins_vs_counts.unstack(level=1).fillna(0)
```

```python
def plot_a_heatmap(dataframe, axis, xlabel, ylabel, title):
    heatmap = axis.pcolor(dataframe, cmap=plt.cm.OrRd)
    # set titles
    axis.set_title(title,y=1.08)
    # turn off the frame, which has a gray background to separate the plots
    axis.set_frame_on(False)
    # center the major ticks at the middle of each cell
    axis.set_yticks(np.arange(dataframe.shape[0])+0.5, minor=False)
    axis.set_xticks(np.arange(dataframe.shape[1])+0.5, minor=False)
    # move x axis labels to the top, and pull it closer to the table
    axis.xaxis.tick_top()
    axis.invert_yaxis()
    # set the labels
    axis.set_xticklabels(xlabel, minor=False);
    axis.set_yticklabels(ylabel, minor=False);
    # turn off ticks
    axis = plt.gca()
    for t in axis.xaxis.get_major_ticks(): 
        t.tick1On = False 
        t.tick2On = False 
    for t in axis.yaxis.get_major_ticks(): 
        t.tick1On = False 
        t.tick2On = False

def plot_n_heatmaps(nrows, ncols, dataframes_list, xlabels, ylabels, titles, figsize):
    assert (nrows*ncols == np.array(map(len, [dataframes_list, xlabels, ylabels, titles]))).all()
    
    fig, ax = plt.subplots(ncols, nrows)
    plt.subplots_adjust(hspace = 0.1)
    fig.set_size_inches(figsize)
            
    for (j, k) in [(x, y) for y in range(ncols) for x in range(nrows)]:
        n = k*(ncols-1) + j
        plot_a_heatmap(dataframes_list[n], ax[k,j], xlabels[n], ylabels[n], titles[n])
```

```python
fig, ax = plt.subplots(figsize=(12,4))
plot_a_heatmap(dayofweek_vs_hour_bins, ax, xlabel=range(1,25), ylabel=day_abbr, title="Call Volume Intensity by Day of Week and Hour of Day")
```

```python
# make x and y labels
hourly_bins_labels_x = range(1, 25)
date_string_labels_y = lambda m : map(lambda d : "{0} {1}".format(day_abbr[d.weekday()], d),
                           call_volume_vs_hour_bins_on_month[m].index.date)
# make titles
cal_title = lambda month_index, year : "{0} {1}".format(month_index, year)
month_years = [cal_title(month_abbr[m], call_volume_vs_hour_bins_on_month[m-1].index.year[0])
               for m in range(1,len(month_abbr))]
# plot 12 heatmaps, one for each month
plot_n_heatmaps(nrows = 3,
               ncols = 4,
               dataframes_list = call_volume_vs_hour_bins_on_month,
               xlabels = [hourly_bins_labels_x for m in range(1,len(month_abbr))],
               ylabels = [date_string_labels_y(m-1) for m in range(1, len(month_abbr))],
               titles = month_years,
               figsize = (19,31))
```
